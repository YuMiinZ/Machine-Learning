```python
condition_null_rent = train_data['v2a1'].isna()

condition_tipovivi3 = train_data['tipovivi3'] == 1

combined_conditions = condition_null_rent & condition_tipovivi3

# Count the number of true instances in the combined condition
count = combined_conditions.sum()
print("La cantidad es: ", count)
```

  Como se observa se puede decir que los valores null representan personas que son dueñas de las casas por lo que los valores se pueden modificar a 0

```python
# Se le aplica tanto al train data como al test data
def set_0(data):
  data.loc[(data['tipovivi1'] == 1), 'v2a1'] = 0
  data.loc[(data['tipovivi4'] == 1), 'v2a1'] = 0
  data.loc[(data['tipovivi5'] == 1), 'v2a1'] = 0

set_0(train_data)
set_0(test_data)

print(train_data)
```

### Miembros de la familia y el valor target

El valor target es asignado por familia por lo que algo a fijarse es si hay miembros de la familia que no comparten el valor. Se observa que si ocurre esta situación por lo que hay que corregirlo.

```python
all_equal = train_data.groupby('idhogar')['Target'].apply(lambda x: x.nunique() == 1)
not_equal = all_equal[all_equal != True]
train_data[train_data['idhogar'] == not_equal.index[0]][['idhogar', 'parentesco1', 'Target']]
```

Para esto se almacena los datos a modificar, esto se hace para saber a que familia pertenece. Lo siguiente es obtener a familia pertenece para obtener el valor del target correcto. Luego nada más se corrige el valor.

```python
def members(data):
  all_equal = data.groupby('idhogar')['Target'].apply(lambda x: x.nunique() == 1)
  not_equal = all_equal[~all_equal]

  for casa in not_equal.index:
      target = int(data.loc[(data['idhogar'] == casa) & (data['parentesco1'] == 1.0), 'Target'].iloc[0])
      data.loc[data['idhogar'] == casa, 'Target'] = target


members(train_data)
#members(test_data)
```

### meaneduc y SQBmeaned
Los datos nulos en las columnas meaneduc y SQBmeaned están relacionados por lo que analizando que representan se puede observar que los valores nulos en meaneduc es de familias que son de una sola persona de 18-19 años. Esto hace que se pueda poner el nivel educativo para quitar los valores nulos.

```python
def mean(data):
  meaneduc_nan=data[data['meaneduc'].isnull()][['Id','idhogar','escolari']]
  me=meaneduc_nan.groupby('idhogar')['escolari'].mean().reset_index()

  for row in meaneduc_nan.iterrows():
    idx=row[0]
    idhogar=row[1]['idhogar']
    m=me[me['idhogar']==idhogar]['escolari'].tolist()[0]
    data.at[idx, 'meaneduc']=m
    data.at[idx, 'SQBmeaned']=m*m

mean(train_data)
mean(test_data)
```

# 4. Fase de modelado:

> Esta fase describe las etapas del entrenamiento y evaluación. Se debe entrenar un único modelo, puede ser de regresión o clasificación.

> Debe constar los diferentes experimentos con sus respectivos parámetros con el fin de seleccionar el mejor modelo. Para ello, no olvide separar los datos en training y testing para la evaluzación de los mismos, y aplicar CrossValidation como técnica para la selección del modelo.

Para este proyecto se utilizará un modelo de clasificación, donde se utilizará los siguientes algoritmos para el entrenamiento de datos:



*   Decision Tree
*   Random Forest
*   K-vecinos más cercanos
*   Naïve Bayes



```python
#Importación de librerías generales
from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score,RepeatedStratifiedKFold
from sklearn.metrics import f1_score, accuracy_score
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

#Árbol de decisión
from sklearn.tree import DecisionTreeClassifier

#Random Forest
from sklearn.ensemble import RandomForestClassifier

# k-Neighbors
from sklearn.neighbors import KNeighborsClassifier

# NB GaussianNB
from sklearn.naive_bayes import GaussianNB

#NB Bernoulli
from sklearn.naive_bayes import BernoulliNB

#División de datos de entrenamiento y el target
x = train_data.drop(columns=['Id', 'idhogar', 'Target'])
y = train_data['Target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

## Decision Tree

```python

# Definir los diferentes hiperparámetros para el árbol de decisión
params = {
    'max_depth': [3, 5, 7, 15, 20, 30],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 7],
}

clf = DecisionTreeClassifier()

# Selección de los mejores parámetros con GridSearchCV
grid_search = GridSearchCV(clf, params, cv=5, scoring='f1_macro')
grid_search.fit(x_train, y_train)
best_params = grid_search.best_params_
best_clf = grid_search.best_estimator_

print("Todos los resultados de la búsqueda:")
means = grid_search.cv_results_['mean_test_score']
stds = grid_search.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, grid_search.cv_results_['params']):
    print(f"Puntuación media: {mean:.3f}, Desviación estándar: {std:.3f} para {params}")

# Entrenamiento del decision tree
y_pred_train = best_clf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred_train)
train_score = f1_score(y_test, y_pred_train, average='macro')

print("\nMejores hiperparámetros:", best_params)
print("Precisión del modelo Decision Tree:", accuracy)
print("Puntuación F1 macro en conjunto de prueba:", train_score)
print("\n Reporte de la clasificación\n",classification_report(y_test, y_pred_train))

```

## Random Forest

```python
params = {

    'n_estimators': [50, 80, 100],
    'max_depth': [3, 5, 7, 10, 30],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 7],
    'bootstrap': [True, False],
}


rf_clf = RandomForestClassifier(n_jobs=-1)


grid_search = GridSearchCV(rf_clf, params, cv=5, scoring='f1_macro')
grid_search.fit(x_train, y_train)
best_params = grid_search.best_params_
best_clf = grid_search.best_estimator_

print("Todos los resultados de la búsqueda para Random Forest:")
means_rf = grid_search.cv_results_['mean_test_score']
stds_rf = grid_search.cv_results_['std_test_score']
for mean_rf, std_rf, params_rf in zip(means_rf, stds_rf, grid_search.cv_results_['params']):
  print(f"Puntuación media: {mean_rf:.3f}, Desviación estándar: {std_rf:.3f} para {params_rf}")


# Entrenamiento y evaluación del modelo
y_pred_train = best_clf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred_train)
train_score = f1_score(y_test, y_pred_train, average='macro')

print("\nMejores hiperparámetros:", best_params)
print("Precisión del modelo Random Forest:", accuracy)
print("Puntuación F1 macro en conjunto de prueba:", train_score)
print("\nReporte de la clasificación\n", classification_report(y_test, y_pred_train))
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib import cm

# Para no modificar los datos originales se copia a una nueva
newData = train_data

# Selecciona las características para el clustering
important_variables = ["v2a1", "r4t3", "tamhog", "escolari", "computer", "qmobilephone"]

# Normaliza las características para que tengan la misma escala
scaler = StandardScaler()
data_scaled = scaler.fit_transform(newData[important_variables])

# Número de clusters (hipotéticamente, podrías experimentar con diferentes valores)
num_clusters = 5

# Aplica K-means para el clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_scaled)

# Graficar los resultados del clustering
plt.figure(figsize=(8, 6))
color_palette = cm.get_cmap('viridis', num_clusters)

for cluster_id in range(num_clusters):
    indices = clusters == cluster_id
    plt.scatter(principal_components[indices, 0], principal_components[indices, 1], color=color_palette(cluster_id), label=f'Cluster {cluster_id}')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering de datos con PCA')
plt.legend()
plt.grid(True)
plt.show()
```
